```
# 必要なライブラリーのダウンロード

import requests
from bs4 import BeautifulSoup
```

```
# 俺速にアクセスし、スクレイピング

r = requests.get("http://jin115.com/")
soup = BeautifulSoup(r.text, features="html.parser")

pages = soup.select("h2 > a")
print(pages)
```

```
# URLを取得

urls = []

for page in pages:
  url = page.get("href")
  urls.append(url)
print(urls)   
```

```
# タイトルを取得

titles = []

for page in pages:
  title = page.get("title")
  titles.append(title)
print(titles)  
```

```
# タイトルとURLを表示する

for title, url in zip(titles, urls):
  print(title, url)
```
